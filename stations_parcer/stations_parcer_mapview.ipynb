## Импорт библиотек

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

## Открытие CSV файла с ж/д

```python
stations_df = pd.read_csv('RU_stations_new.csv', sep='\t', encoding='utf-8')
# stations_df = stations_df[stations_df['railway'] == 'Свердловская']
stations_df
```

## Визуализация по территориальным участкам

```python
show = 'railway' # отображение ж/д по принадлежности к территориальному участку

fig = px.scatter_mapbox(stations_df, lat='lat', lon='lon', color=show, height = 700, text='name',
                        mapbox_style="open-street-map")
fig
```

## Визуализация по областям

```python
show = 'region' # отображение ж/д по принадлежности к области

fig = px.scatter_mapbox(stations_df, lat='lat', lon='lon', color=show, height = 700, text='name',
                        mapbox_style="open-street-map")
fig
```
